```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
import json
data = json.loads(response)

for d in data:
    print(d['title'])
```

```python
print(len(data))
```

```python
for i in data:
    print(i['title'])
    print(i['createdAt'])
    print(i['commentCount'])
    print(i['likeCount'])
```

```python
like=0
comment=0

for i in data:
    comment =comment+ i['commentCount']
    like =like + i['likeCount']
    
print(int(comment/len(data)), int(like/len(data)))
```
